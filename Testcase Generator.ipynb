```python
import random

def generate_dataset(n, capacity, max_time):
    filename = f"input{n}.txt"

    with open(filename, "w") as f:
        f.write(f"{n} {capacity}\n")

        for _ in range(n):
            price = random.randint(5, 20)
            demand = random.randint(3, 12)
            l = random.randint(1, max_time - 6)
            r = l + random.randint(2, 6)

            f.write(f"{price} {demand} {l} {r}\n")

    print(f"Written to {filename}")


generate_dataset(10,   capacity=20,   max_time=100)
generate_dataset(25,   capacity=50,   max_time=250)
generate_dataset(50,   capacity=100,  max_time=500)
generate_dataset(100,  capacity=200,  max_time=1000)
generate_dataset(200,  capacity=400,  max_time=1000)
generate_dataset(250,  capacity=500,  max_time=1000)
generate_dataset(500,  capacity=1000, max_time=1000)
generate_dataset(1000, capacity=2000, max_time=1000)
generate_dataset(2000, capacity=4000, max_time=1000)
```
